# read_file()
this method to read any dataset such as 'CSV', 'Excel' and 'jason'.\
The method also detects the file extension automatically

Syntax:
- dpk.read_file(file_path)

Parameters:
- file_path : location of the jason, csv or excel file

Returns:
- the data in the form of the data frame

```python
df = dpk.read_file('dataset1.json')
df.head()
```

```python
df = dpk.read_file('dataset3.xlsx')
df.head()
```

```python
df = dpk.read_file('dataset2.csv')
df.head()
```

# data_summary()
this method to summarize the data frame by displaying the:
- count : number of the non missing data for each feature
- missing val : number of the missing data for each feature
- missing % : percentage of the missing data for each feature
- unique : number of the uniqe data for each feature
- most freq : shows you the most frequent data for each feature
- most freq cnt : shows you how many the most frequent data are repeated for each feature
- mean : calculate the mean for the numerical feature and if the feature is categorical it is display 'NaN'
- data type : shows you the data type for each feature

Syntax:
- dpk.data_summary(data_frame)

Parameters:
- data_frame : the data frame you want to summarize it

Returns:
- data frame contains the count, missing value, percentage of the missing value, uniqe, most frequent data, count of the most frequent data, mean and data type for all features

```python
dpk.data_summary(df)
```

# row_missing_data_summary()
this method shows you how many rows contain missing data and they are display it into 5 categories:
- 0% : the number of rows that don't contain missing data
- 1% -> 25% : the number of rows that contain missing data between 1% and 25% from the total number of features in one row
- 25% -> 50% : the number of rows that contain missing data between 25% and 50% from the total number of features in one row
- 50% -> 75% : the number of rows that contain missing data between 50% and 75% from the total number of features in one row
- 75% -> 100% : the number of rows that contain missing data between 75% and 100% from the total number of features in one row

Syntax:
- dpk.row_missing_data_summary(data_frame)

Parameters:
- data_frame : the data frame you want to know how many rows contain missing data

Returns:
- 5 categories\
    0%\
    1% -> 25%\
    25% -> 50%\
    50% -> 75%\
    and 75% -> 100% in the form of the data frame

```python
dpk.row_missing_data_summary(df)
```

# drop_columns()
this method is to drop any number of columns or features you want to drop 

Syntax:
- dpk.drop_columns(data_frame, columns_list)

Parameters:
- data_frame : the data frame
- columns_list : you put the columns names you want to drop it in list

Returns:
- the data frame without the columns you dropped it

```python
columns_list = ['installer', 'scheme_name']
new_df = dpk.drop_columns(df, columns_list)
new_df.head()
```

# drop_col_contains_missing_percentage()
this method is to drop all columns or features that contain more than the percentage of missing data that you specify

Syntax:
- dpk.drop_col_contains_missing_percentage(data_frame, percentage)

Parameters:
- data_frame : the data frame
- percentage : this is the percentage you specify to drop all columns that contain more than this percentage of missing data

Returns:
- the data frame without the columns that contain more than the percentage of missing data

```python
# the percentage = 5%
new_df = dpk.drop_col_contains_missing_percentage(df, 0.05)
new_df.head()
```

# drop_row_contains_missing_percentage()

# mean_imputation()
this method is to fill all the missing value in numerical features with the mean of the numerical feature

Syntax:
- dpk.mean_imputation(data_frame_column)

Parameters:
- data_frame_column : the numerical column or feature that contain of missing value

Returns:
- the numerical column filled with the mean in the missing value

```python
new_df = df.copy()
new_df['amount_tsh'] = dpk.mean_imputation(new_df['amount_tsh'])
dpk.data_summary(new_df)
```

# mode_imputation()

# forward_fill_imputation()
a method of handling missing or incomplete data by carrying forward the last observed value to fill in the gaps

Syntax:
- dpk.forward_fill_imputation(data_frame_column)

Parameters:
- data_frame_column : the column or feature that contain of missing value

Returns:
- the column filled with the last observation carried forward in the missing value

```python
new_df['funder'] = dpk.forward_fill_imputation(new_df['funder'])
```

# backward_fill_imputation()
a method of handling missing or incomplete data by fill missing values with next data point

Syntax:
- dpk.backward_fill_imputation(data_frame_column)

Parameters:
- data_frame_column : the column or feature that contain of missing value

Returns:
- the column filled with missing values with next data point

```python
new_df['funder'] = dpk.backward_fill_imputation(new_df['funder'])
```
